```python
def get_event(url):
    next_page = url
    while True:
        response = requests.get(next_page)
        data = response.json()
        for event in data['results']:
            yield event
        next_page = data['next']
        if not next_page or next_page == "None":
            break
    return events
```

```python
type_counter = {}
tag_counter = {}
for event in all_events:
    for tag in event['tags']:
        if tag not in tag_counter:
            tag_counter[tag] = 0
        tag_counter[tag] += 1
    for event_type in event['categories']:
        if event_type not in type_counter:
            type_counter[event_type] = 0
        type_counter[event_type] += 1
```

```python
def tag_to_category(tag, category):
    for event in all_events:
        if tag in event['tags'] and category not in event['categories']:
            event['categories'].append(category)
            del event['tags'][event['tags'].index(tag)]
            
def category_to_tag(tag, category):
    for event in all_events:
        if category in event['categories'] and tag not in event['tags']:
            event['tags'].append(tag)
            del event['categories'][event['categories'].index(category)]
            
def check_tag_category(tag, category):
    return len([event for event in all_events if tag in event['tags'] and category not in event['categories']])

def remove_tag(tag):
    for event in all_events:
        if tag in event['tags']:
            del event['tags'][event['tags'].index(tag)]
            
def remove_category(category):
    for event in all_events:
        if category in event['categories']:
            del event['categories'][event['categories'].index(category)]
            
def tag_to_tags(source_tag, target_tags):
    for event in all_events:
        if source_tag in event['tags']:
            del event['tags'][event['tags'].index(source_tag)]
            for target_tag in target_tags:
                if target_tag not in event['tags']:
                    event['tags'].append(target_tag)

def tag_to_tags_and_cat(source_tag, target_tags, target_cats):
    for event in all_events:
        if source_tag in event['tags']:
            del event['tags'][event['tags'].index(source_tag)]
            for target_tag in target_tags:
                if target_tag not in event['tags']:
                    event['tags'].append(target_tag)
            for cat in target_cats:
                if cat not in event['categories']:
                    event['categories'].append(cat)
```

```python
def write_events(event_tags, output_type, is_tag=True, stop_words=[], stop_tags=[], stop_types=[]):
    if is_tag:
        tag_key = 'tags'
    else:
        tag_key = 'categories'
    with open(output_type + ".json", "w+", encoding="utf-8") as f:
        events = [
            {"title": e['description'], "description": e['body_text'], "tags": e['tags'], "type": [output_type], "types": e['categories']} 
            for e in all_events 
            if any(t in e[tag_key] for t in event_tags) 
            and all(s not in e['description'] and s not in e['body_text'] for s in stop_words)
            and all(t not in e['tags'] for t in stop_tags)
            and all(t not in e['categories'] for t in stop_types)
        ]
        json.dump(events, f, ensure_ascii=False, indent=4)
```

```python
def write_events_2(output_type, predicate):
    with open(output_type + ".json", "w+", encoding="utf-8") as f:
        events = [
            {"title": e['description'], "description": e['body_text'], "tags": e['tags'], "type": [output_type], "types": e['categories']} 
            for e in all_events 
            if predicate(e)
        ]
        json.dump(events, f, ensure_ascii=False, indent=4)
```

```python
def is_sport(e):
    return any(t in e['tags'] for t in sport_tags) \
        and all(t not in e['categories'] for t in ["party"])

def is_training(e):
    return any(t in e['tags'] for t in training_tags) \
        and all(s not in e['description'] and s not in e['body_text'] for s in ["лекц"]) \
        and all(t not in e['categories'] for t in ["party", "yarmarki-razvlecheniya-yarmarki", "festival", "stock"])

def is_lecture(e):
    return any(t in e['tags'] for t in lecture_tags) \
        and all(s not in e['description'] and s not in e['body_text'] for s in ["мастер-класс", "тренинг", "семинар", "ручн", "урок", "Форум", "форум", "курс"]) \
        and all(t not in e['categories'] for t in ["festival", "party", "yarmarki-razvlecheniya-yarmarki", "exhibition", "stock"]) \
        and all(t not in e['tags'] for t in ["парки аттракционов", "бизнес-образование", "кино (развлечения)", "квесты и игры"] + training_tags)

def is_game(e):
    return any(t in e['tags'] for t in quest_tags) \
        and all(t not in e['categories'] for t in ["festival", "party"]) \
        and all(t not in e['tags'] for t in ["парки аттракционов", "бизнес-образование"])

def is_view(e):
    return any(t in e['tags'] for t in view_tags) \
        and all(t not in e['categories'] for t in ["festival", "party"]) \
        and all(t not in e['tags'] for t in ["парки аттракционов", "бизнес-образование"])

def is_food(e):
    return any(t in e['tags'] for t in food_tags) \
        and all(s not in e['description'] and s not in e['body_text'] for s in ["лекци", "лекто", "квиз"]) \
        and all(t not in e['categories'] for t in ["party", "concert", "exhibition", "cinema"]) \
        and all(t not in e['tags'] for t in training_tags + lecture_tags)

def is_party(e):
    return any(t in e['categories'] for t in party_tags) \
        and all(s not in e['description'] and s not in e['body_text'] for s in ["лекци", "лекто", "квиз"]) \
        and all(t not in e['categories'] for t in ["concert", "festival", "kids", "yarmarki-razvlecheniya-yarmarki", "tour"]) \
        and all(t not in e['tags'] for t in training_tags + lecture_tags + ["ночь музеев"])

def is_theater(e):
    return any(t in e['categories'] for t in theatre_tags) \
        and all(s not in e['description'] and s not in e['body_text'] for s in ["лекци", "лекто", "квиз"]) \
        and all(t not in e['categories'] for t in ["concert", "festival", "kids", "yarmarki-razvlecheniya-yarmarki", "tour", "party", "exhibition"]) \
        and all(t not in e['tags'] for t in training_tags + lecture_tags + ["ночь музеев"])    

def is_concert_programm(e):
    return any(t in e['tags'] for t in concert_programm_tags) \
        and all(s not in e['description'] and s not in e['body_text'] for s in ["лекци", "лекто", "квиз"]) \
        and all(t not in e['categories'] for t in ["yarmarki-razvlecheniya-yarmarki", "tour", "theater", "festival"]) \
        and all(t not in e['tags'] for t in training_tags + lecture_tags + ["ночь музеев"])      

def is_standup(e):
    return any(t in e['tags'] for t in standup_tags) \
        and all(s not in e['description'] and s not in e['body_text'] for s in ["лекци", "лекто", "квиз"]) \
        and all(t not in e['categories'] for t in ["yarmarki-razvlecheniya-yarmarki", "tour", "theater", "festival"]) \
        and all(t not in e['tags'] for t in training_tags + lecture_tags + ["ночь музеев"])

def is_show(e):
    return any(t in e['tags'] for t in show_tags) \
        and all(s not in e['description'] and s not in e['body_text'] for s in ["лекци", "лекто", "квиз"]) \
        and all(t not in e['categories'] for t in ["festival", "yarmarki-razvlecheniya-yarmarki", "tour", "party", "exhibition", "cinema"]) \
        and all(t not in e['tags'] for t in training_tags + lecture_tags + standup_tags)    

def is_market(e):
    return any(t in e['categories'] for t in market_tags) \
        and all(t not in e['categories'] for t in ["tour", "party", "exhibition", "cinema", "social-activity"])

def is_tour(e):
    return any(t in e['categories'] for t in tour_tags) \
        and all(t not in e['categories'] for t in ["party", "cinema", "yarmarki-razvlecheniya-yarmarki", "exhibition"])

def is_exhibition(e):
    return any(t in e['categories'] for t in exhibition_tags) \
        and all(t not in e['categories'] for t in ["party", "cinema", "yarmarki-razvlecheniya-yarmarki", "tour"])  

def is_festival(e):
    return any(t in e['categories'] for t in festival_tags) \
        and all(t not in e['categories'] for t in ["yarmarki-razvlecheniya-yarmarki", "tour"]) \
        and all(t not in e['tags'] for t in ["флешмобы"] + sport_tags)

def is_compete(e):
    return any(t in e['tags'] for t in compete_tags)

def is_concert(e):
    return len(e['categories']) == 1 and e['categories'][0] == 'concert'and any(t in e['tags'] for t in concert_tags)
```

```python
type_predicates = {
    "sport": is_sport,
    "training": is_training,
    "lecture": is_lecture,
    "game": is_game,
    "view": is_view,
    "food": is_food,
    "party": is_party,
    "theatre": is_theater,
    "concert_programm": is_concert_programm,
    "standup": is_standup,
    "show": is_show,
    "market": is_market,
    "tour": is_tour,
    "exhibition": is_exhibition,
    "festival": is_festival,
    "compete": is_compete,
    "concert": is_concert
}
```

```python
preserve_tags = ["online", "free", "open air"]
output_events = []
for event in all_events:
    output_tags = []
    for event_type, predicate in type_predicates.items():
        if predicate(event):
            output_tags.append(event_type)
    
    if len(output_tags) == 0:
        continue
    
    for t in preserve_tags:
        if t in event['tags']:
            output_tags.append(t)
            
    output_event = {
        "title": event['description'],
        "description": event['body_text'],
        "types": output_tags
    }
    output_events.append(output_event)
```
